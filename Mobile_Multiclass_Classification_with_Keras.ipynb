## Mobile Price Range Multiclass Classification

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

## Read the train test data

```python
import pandas as pd
df=pd.read_csv('/content/train_mobile.csv')
df.head()
```

### Perform data quality check

```python
df.info()
```

```python
df.isna().sum()
```

### Seperate X and Y (price_range)

```python
X=df.drop(columns=['price_range'])
Y=df[['price_range']]
```

```python
X.head()
```

```python
Y.head()
```

Price range

0 - Low cost

1 - Medium cost

2 - High cost

3 - Very High cost

```python
Y.value_counts()
```

### Preprocess the data on X

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy='median')),
                         ('scaler',StandardScaler())]).set_output(transform='pandas')
```

```python
X_pre=num_pipe.fit_transform(X)
X_pre.head()
```

### Apply Train test split

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X_pre,Y,test_size=0.2,random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
xtrain.shape[1]
```

## For multiclass classification one hot encode the target features

```python
from keras.utils import to_categorical
ytrain_cat=to_categorical(ytrain)
ytest_cat=to_categorical(ytest)
```

```python
ytrain_cat[0:5]
```

```python
ytrain.shape
```

```python
ytest_cat[0:5]
```

```python
ytest.head()
```

```python
df['price_range'].unique()
```

### Build the neural network

# image.png

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.callbacks import EarlyStopping
```

```python
model=Sequential()

# create Dense Layer 1
model.add(Dense(units=64,activation='relu',input_shape=(xtrain.shape[1],)))

# Add dropout Layer 1
model.add(Dropout(rate=0.3))

# Create Dense Layer 2
model.add(Dense(units=32,activation='relu'))

# Add droput Layer 2
model.add(Dropout(rate=0.2))

# output layer haev 4 neuros because there are 4 classes in this data
model.add(Dense(units=4,activation='softmax'))
```

```python
model.summary()
```

### Complile the model

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

### Add early stopping

# image.png

```python
es=EarlyStopping(monitor='val_loss',patience=10)
```

### Traning the model

```python
hist=model.fit(xtrain,ytrain_cat,validation_split=0.2,epochs=200,callbacks=[es])
```

### Plot the curve

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.plot(hist.history['loss'],label='Traning Loss')
plt.plot(hist.history['val_loss'],label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Categorical cross entropy')
plt.title('Learning Curve')
plt.legend()
plt.show()
```

### Evaluate the model

```python
model.evaluate(xtrain,ytrain_cat)
```

```python
model.evaluate(xtest,ytest_cat)
```

### Get probability for train and test

```python
yprob_train=model.predict(xtrain)
yprob_test=model.predict(xtest)
```

```python
yprob_train[0:5]
```

```python
yprob_test[0:5]
```

### Get Predictions out of probability

```python
import numpy as np
ypred_train=np.argmax(yprob_train,axis=1)
ypred_test=np.argmax(yprob_test,axis=1)
```

```python
ypred_train[0:5]
```

```python
ytrain.head()
```

```python
ypred_test[0:5]
```

```python
ytest.head()
```

## Plot confusion matrix

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cf=confusion_matrix(ytest,ypred_test)
cfd=ConfusionMatrixDisplay(cf)
cfd.plot()
```

## Printing classification report

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred_test))
```
